```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
from  dash import Dash, dcc, html, callback, Output, Input
#import dash_design_kit as ddk
```

```python
df = pd.read_csv('gcpEconomicSectors.csv')
df.head()
```

```python
app= Dash(__name__)
```

```python
# Line Graph of GCP vs Year, with a dropdown County
app.layout = html.Div([
    html.H1(children='KENYA GROSS COUNTY PRODUCT(2018-2020)',style={'textAlign':'center'}),
    dcc.Dropdown(df.County.unique(), 'Baringo', id='dropdown-selection'),
    dcc.Graph(id='graph-content')
])

@callback(
    Output('graph-content','figure'),
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = df[df.County==value]
    return px.line(dff, x='Year', y='GCP')
```

```python
fig = px.bar(df.County)
app.layout = html.Div(children=[
    html.Div(children='''Bar Graph''', style={'textAlign':'center'}),
    dcc.Graph(id='example-graph',
              figure=fig
    )
    
])
```

```python
if __name__ == '__main__':
    app.run_server(debug=True, jupyter_mode='tab')
```
